```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
```

Sequential: một lớp mô hình trong TensorFlow để xây dựng các mô hình mạng nơ-ron tuần tự (sequential neural networks), tức là các lớp được xếp chồng lên nhau theo một thứ tự nhất định.

Dense: một lớp trong TensorFlow để thêm các lớp đầy đủ kết nối (fully connected layers) vào mô hình mạng nơ-ron.

SGD và Adam: hai tối ưu hóa (optimization) được sử dụng để tối ưu hóa hàm mất mát (loss function) trong quá trình huấn luyện mô hình mạng nơ-ron.

EarlyStopping: một callback trong TensorFlow để dừng quá trình huấn luyện mô hình sớm nếu việc huấn luyện không cải thiện độ chính xác (accuracy) của mô hình trong một số vòng lặp.


```python
df = pd.read_csv("fetal_health.csv")
df.head()
```

```python
df.info()
```

```python
scaler = MinMaxScaler()
```

```python
X = df.drop("fetal_health", axis = 1)
y = df[["fetal_health"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                   random_state=6
                                                   )
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
model = Sequential()
model.add(Dense(64, input_shape = (21, ), activation = 'relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='softmax'))
model.compile(optimizer='Adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, y_train, epochs=150, validation_split=0.1)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print("Loss: " + str(loss));
print("Accuracy: " + str(accuracy));
```
